Finding the distance using geodesic in kms

```python
cols=['Restaurant_latitude','Restaurant_longitude','Delivery_location_latitude','Delivery_location_longitude']
for col in cols:
    df[col]= abs(df[col])
```

```python
import numpy as np
from geopy.distance import geodesic

df['distance']=np.zeros(len(df))
restaurant_cordinates_df=df[['Restaurant_latitude','Restaurant_longitude']].to_numpy()
delivery_location_cordinates_df=df[['Delivery_location_latitude','Delivery_location_longitude']].to_numpy()

for i in range(len(df)):
    df['distance'].loc[i]=geodesic(restaurant_cordinates_df[i],delivery_location_cordinates_df[i])
```

```python
df['distance'] = df['distance'].astype('str').str.extract('(\d+)')
```

```python
df['distance'] = df['distance'].astype("float64")
df['distance']
```

```python
df['distance'].describe()
```

Finding the Order preparation time in minutes

```python
df['Time_Orderd'].unique()
```

```python
df['Time_Order_picked'].unique()
```

```python
def filter_int_float_values(column_df):
    for i in range(len(column_df)):
        if '.' in str(column_df.iloc[i]) or  column_df.iloc[i] == '1':
            column_df.iloc[i] = np.nan

    return column_df
df['Time_Orderd'] = filter_int_float_values(df['Time_Orderd']).str.slice(0,5)
df['Time_Order_picked'] = filter_int_float_values(df['Time_Order_picked']).str.slice(0,5)
```

```python
df['Time_Orderd'].unique()
```

```python
df['Time_Order_picked'].unique()
```

there has been nan values in those column

```python
df['Time_Orderd'].fillna(df['Time_Orderd'].mode()[0],inplace = True)
df['Time_Order_picked'].fillna(df['Time_Order_picked'].mode()[0], inplace = True)
```

Concatenating :00 atlast to make it date format and also we can calculate the preparation time using these values

```python
df['Time_Orderd'] = df['Time_Orderd'] + ':00'
df['Time_Order_picked'] = df['Time_Order_picked'] + ':00'
df = df.reset_index()
df.drop(columns=['index'],axis=1,inplace=True)
```

```python
df['Time_Order_picked']
```

```python
df.head()
```

```python
df['Time_Orderd']=pd.to_timedelta(df['Time_Orderd'])
df['Time_Order_picked']=pd.to_timedelta(df['Time_Order_picked'])

df['Time_Order_picked_formatted']=np.where(df['Time_Order_picked'] < df['Time_Orderd'], df['Order_Date'] + pd.DateOffset(1)+df['Time_Order_picked'], df['Order_Date']+df['Time_Order_picked'])
df['Time_Ordered_formatted'] = df['Order_Date']+ df['Time_Orderd']
df['order_preparation_time']=((df['Time_Order_picked_formatted']- df['Time_Ordered_formatted']).dt.total_seconds())/60
```

```python
df['order_preparation_time'].value_counts()
```

```python
df.isnull().sum()
```

```python
df['Delivery_person_Age'].fillna(np.mean(df['Delivery_person_Age']),inplace =True)
```

```python
df['Delivery_person_Ratings'].fillna(np.mean(df['Delivery_person_Ratings']),inplace =True)
```

```python
df['Weather_conditions'].fillna(df['Weather_conditions'].mode()[0],inplace=True)
df['Road_traffic_density'].fillna(df['Road_traffic_density'].mode()[0],inplace=True)
df['multiple_deliveries'].fillna(np.mean(df['multiple_deliveries']),inplace=True)
df['Festival'] = df['Festival'].fillna('No')
df['City'].fillna(df['City'].mode()[0],inplace=True)
```
